# My Question
###What is the probability that a Wikipedia article has a "See also" section specifically for animals?

# My Answer

```python
import requests
from bs4 import BeautifulSoup

# URL of the Wikipedia article on Chicken
url = "https://en.wikipedia.org/wiki/Chicken"

# Send an HTTP request to the URL and get the HTML content of the page
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Search for the "See also" section by looking for an id or heading
    see_also_section = soup.find('span', {'id': 'See_also'})
    
    # If the "See also" section exists, return True, otherwise False
    if see_also_section:
        print("The article has a 'See also' section.")
        probability = 1  # Probability is 1 if the section exists
    else:
        print("The article does not have a 'See also' section.")
        probability = 0  # Probability is 0 if the section does not exist

else:
    print("Failed to retrieve the page.")
    probability = None  # Could not retrieve page

# Display the probability
print(f"Probability that the article has a 'See also' section: {probability}")
```

```python
import requests
from bs4 import BeautifulSoup

# List of URLs to scrape
urls = [
    "https://en.wikipedia.org/wiki/Dog",
    "https://en.wikipedia.org/wiki/Cat",
    "https://en.wikipedia.org/wiki/Frog",
    "https://en.wikipedia.org/wiki/Zebra",
    "https://en.wikipedia.org/wiki/Giraffe",
    "https://en.wikipedia.org/wiki/Bird",
    "https://en.wikipedia.org/wiki/Turtle",
    "https://en.wikipedia.org/wiki/Pig",
    "https://en.wikipedia.org/wiki/Sheep",
    "https://en.wikipedia.org/wiki/Cow"
]

# Function to check if the article has a "See also" section
def check_see_also(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        see_also_section = soup.find('span', {'id': 'See_also'})
        return 1 if see_also_section else 0
    else:
        return None  # If the page cannot be retrieved

# Loop through each URL and check for "See also"
results = {}
for url in urls:
    probability = check_see_also(url)
    results[url] = probability
    # Print the results for each article
    print(f"URL: {url} - Probability of 'See also' section: {probability}")

# If you want to see the final results stored in a dictionary
print("\nFinal Results:")
for url, probability in results.items():
    print(f"{url} -> Probability: {probability}")
```
